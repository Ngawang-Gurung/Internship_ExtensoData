```python
import pandas as pd
```

```python
df= pd.read_csv('transaction_summary.csv', delimiter=',')
df.head()
```

```python
df[['WITHDRAWAL AMT', 'DEPOSIT AMT', 'BALANCE AMT']] = df[['WITHDRAWAL AMT', 'DEPOSIT AMT', 'BALANCE AMT']].fillna(value = 0)
```

```python
df['balance'] = (df['DEPOSIT AMT'] - df['WITHDRAWAL AMT']).cumsum()
```
